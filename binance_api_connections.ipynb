```python
from binance.client import Client
import pandas as pd
```

```python
api_key = ''
secret_key = ''
```

```python
client = Client(api_key, secret_key)
```

```python
depth = client.get_order_book(symbol = 'BTCUSDT', limit = 500)
```

```python
df_depth = pd.DataFrame(depth)
```

```python
def col_separator (df):
    df['bid_price'] = df['bids'].apply(lambda x: x[0])
    df['ask_price'] = df['asks'].apply(lambda x: x[0])
    df['bid_volume'] = df['bids'].apply(lambda x: x[1])
    df['ask_volume'] = df['asks'].apply(lambda x: x[1])
    return df
```

```python
df_depth = col_separator(df_depth).drop(columns = ['bids','asks'])
df_depth
```

```python
df_depth.info()
```

```python
df_depth['bid_price'] = df_depth['bid_price'].astype('float')
df_depth['ask_price'] = df_depth['ask_price'].astype('float')
df_depth['bid_volume'] = df_depth['bid_volume'].astype('float')
df_depth['ask_volume'] = df_depth['ask_volume'].astype('float')
```

```python
bid_vol = df_depth['bid_volume'].sum()
ask_vol = df_depth['ask_volume'].sum()

print('Bid total volume: {}'.format(bid_vol))
print('Ask total volume: {}'.format(ask_vol))

```

```python
trades = client.get_recent_trades(symbol = 'BTCUSDT', limit = 1000)
```

```python
df_trades = pd.DataFrame(trades)
df_trades
```

```python
df_trades['time'] = pd.to_datetime(df_trades['time'], unit = 'ms')
df_trades['price'] = df_trades['price'].astype('float')
df_trades['qty'] = df_trades['qty'].astype('float')
df_trades['quoteQty'] = df_trades['quoteQty'].astype('float')

df_trades
```

```python
df_trades.groupby('isBuyerMaker')['quoteQty'].sum().reset_index()
```
