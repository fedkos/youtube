```python
import pandas as pd
import numpy as np
from binance.client import Client
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
api_key = ''
secret_key = ''
```

```python
client = Client(api_key, secret_key)
```

```python
start = '2022-01-01'
end = '2022-03-01'
```

```python
klines = client.get_historical_klines('BTCUSDT', Client.KLINE_INTERVAL_1HOUR, start, end)
```

```python
df = pd.DataFrame(klines)
df.columns = ['open_time', 'open', 'high', 'low', 'close', 'volume', 
              'close_time', 'q_vol', 'trades', 'taker_buy_volume', 'taker_q_volume', 'ignored']

df['open_time'] = pd.to_datetime(df['open_time'], unit = 'ms')
df['close_time'] = pd.to_datetime(df['close_time'], unit = 'ms')

df[['open', 'high', 'low', 'close',
    'volume', 'q_vol', 'trades', 'taker_buy_volume', 'taker_q_volume']] = df[['open', 'high', 'low', 'close', 'volume', 'q_vol', 'trades', 'taker_buy_volume', 'taker_q_volume']].astype('float')
```

```python
df = df.tail(150)
```

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(go.Candlestick(x = df['open_time'],
                            open = df['open'],
                            high = df['high'],
                            low = df['low'],
                            close = df['close'],
                            name = 'BTCUSDT'),
                            row = 1, col = 1)

fig.add_trace(go.Bar(
            x = df['open_time'],
            y = df['volume'],
            name = 'Volume',
            marker_color = 'black', opacity = 0.4), row = 1, col = 1, secondary_y = True)

fig.add_trace(go.Bar(
            x = df['open_time'],
            y = df['taker_buy_volume'],
            name = 'Volume',
            marker_color = 'red', opacity = 0.7), row = 1, col = 1, secondary_y = True)

fig.update_layout(xaxis_rangeslider_visible = False)

fig.show()

```
